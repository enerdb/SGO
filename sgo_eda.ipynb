# Cleaning functions

```python
def string_ate_ponto(string):
    return string.split('.')[0]

def arruma_data_hora(df, coluna):
    df[coluna] = df[coluna].apply(string_ate_ponto)
    df[coluna] = pd.to_datetime(df[coluna], format = '%d/%m/%Y %H:%M:%S')
    return df
    
```
